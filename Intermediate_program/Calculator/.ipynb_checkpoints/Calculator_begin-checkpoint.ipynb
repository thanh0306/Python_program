## 1. First calculator program
- Execute basic calculator
- Input two number by user and excute plus, minus, multiplication, division

```python
print("Calculator")
number_1 = int(input("Give the first number: "))
number_2 = int(input("Give the second number: "))
print("(1) +\n(2) -\n(3) *\n(4) /")
number_3 = int(input("Please select something (1-4): "))
if number_3 == 1:
    result = number_1 + number_2
    print("The result is: {}".format(result))
elif number_3 == 2:
    result = number_1 - number_2
    print("The result is: {}".format(result))
elif number_3 == 3:
    result = number_1 * number_2
    print("The result is: {}".format(result))
elif number_3 == 4:
    result = float(number_1 / number_2)
    print("The result is: {}".format(result))
else: 
    print("Selection was not correct.")
```

## 2. Calculator second
- Several options, changing the number

```python
print("Calculator")
result = 0
select = 0 
number1 = int(input("Give the first number: "))
number2 = int(input("Give the second number: "))
while select != 6:
    print("(1) +\n(2) -\n(3) *\n(4) /\n(5) Change numbers\n(6) Quit")
    print("Current numbers: {} {}".format(number1, number2))
    select = int(input("Please select something (1-6): "))
    if select == 1:
        result = number1 + number2
        print("The result is: {}".format(result))
    elif select == 2:
        result = number1 - number2
        print("The result is: {}".format(result))
    elif select == 3:
        result = number1 * number2
        print("The result is: {}".format(result))
    elif select == 4:
        result = number1 / number2
        print("The result is: {}".format(result))
    elif select == 5:
        number1 = int(input("Give the first number: "))
        number2 = int(input("Give the second number: "))
    continue  
                   
print("Thank you!") 
```

## 3. Calculator third step
The same as second calculator but adding more 2 function which can calculate sin and cos number 1 and number 2

```python
import math 
print("Calculator")
result = 0
select = 0 
number1 = int(input("Give the first number: "))
number2 = int(input("Give the second number: "))
while select != 8:
    print("(1) +\n(2) -\n(3) *\n(4) /\n(5) sin(number1/number2)\n(6) cos(number1/number2)\n(7) Change numbers\n(8) Quit")
    print("Current numbers: {} {}".format(number1, number2))
    select = int(input("Please select something (1-8): "))
    if select == 1:
        result = number1 + number2
        print("The result is: {}".format(result))
    elif select == 2:
        result = number1 - number2
        print("The result is: {}".format(result))
    elif select == 3:
        result = number1 * number2
        print("The result is: {}".format(result))
    elif select == 4:
        result = number1 / number2
        print("The result is: {}".format(result))
    elif select == 5:
        temp = number1 / number2
        result = math.sin(temp)
        print("The result is: {}".format(result))
    elif select == 6:
        temp = number1 / number2
        result = math.cos(number1/number2)
        print("The result is: {}".format(result))
    elif select == 7:
        number1 = int(input("Give the first number: "))
        number2 = int(input("Give the second number: "))
    continue  
                   
print("Thank you!") 
```
