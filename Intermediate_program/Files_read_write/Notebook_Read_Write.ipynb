## 1. Notebook: Read and write to the notebook - Begin
Create a program which allows the user to
- (1) Read the contents of the notebook
- (2) Add notes to the file or
- (3) Delete all of the notes.
- If the user selects 1, the entire notebook file is printed to the screen, if 2 then the program prompts "Write a new note: ", and adds the written note as the last line into the file with a trailing line break "\n". If the player selects 3, the file is emptied and the message "Notes deleted" will be shown. Also add the option (4) Quit, which ends the program, printing "Notebook shutting down, thank you.". With other selections the program prompts "Incorrect selection".

```python
select = 0

while select != 4:
    print("(1) Read the notebook\n(2) Add note\n(3) Empty the notebook\n(4) Quit\n")
    select = int(input("Please select one: "))
    if select == 1:
        with open("notebook.txt", "r") as f:
            content = f.read()
            print(content)
    elif select == 2:
        with open("notebook.txt", "a") as f:
            content = input("Write a new note: ")
            f.write(content+"\n")
    elif select == 3:
        with open("notebook.txt", "w") as f:
            content = ""
            f.write(content)
            print("Notes deleted.")
print("Notebook shutting down, thank you.")
```

## 2. Adding time to notebook
The same function as the first notebook but in this second notebook program, I will update small feature the date and time of the written note to the program. 

```python
import time
select = 0
while select != 4:
    print("(1) Read the notebook\n(2) Add note\n(3) Empty the notebook\n(4) Quit\n")
    select = int(input("Please select one: "))
    if select == 1:
        with open("notebook.txt", "r") as f:
            content = f.read()
            print(content)
    elif select == 2:
        with open("notebook.txt", "a") as f:
            content = input("Write a new note: ")
            f.write(content+":::"+time.strftime("%X %x")+"\n")
    elif select == 3:
        with open("notebook.txt", "w") as f:
            content = ""
            f.write(content)
            print("Notes deleted.")
print("Notebook shutting down, thank you.")
```

## 3. Notebook with exception handler
- Also the other continuous project, the notebook, has relied on user actions in the sense that it would have broken down if the user had decided to read the file without writing anything to it. In this exercise we fix this, and add the possiblity of changing the used notebook file while the program is running.


- First of all, make the program start by checking if there is a file "notebook.txt" and create one if there is none. If this has to be done, also inform the user with the warning "No default notebook was found, created one.".


- Adding a fourth selection to the notebook, "(4) Change the notebook". If the user selects this option, the user is prompted for a new file "Give the name of the new file: ". If there is an existing file, it is opened and loaded into the notebook program, while the old notebook file is closed. If the new notebook file does not exist, the system informs the user "No notebook with that name detected, created one." and makes a new file. Also add a note of the used notebook file to the main menu, "Now using file [filename]".

```python
import time
def getfile(name):
    try:
        with open(name, "r") as f:
            content = f.read()
    except FileNotFoundError:
        with open(name, "w") as f:
            
            print("No notebook with that name detected, created one.")
    



def main():
    default = "notebook.txt"
    try:
        with open(default, "r") as f:
            
            content = f.read()                       
    except FileNotFoundError:
        with open(default, "w") as f:
            
            print("No default notebook was found, created one.")
            
    
    
    select = 0
    while select != 5:
        print("Now using file {}".format(default))
        print("(1) Read the notebook\n(2) Add note\n(3) Empty the notebook\n(4) Change the notebook\n(5) Quit\n")
        select = int(input("Please select one: "))
        if select == 1:
            with open(default, "r") as f:
                content = f.read()
                print(content)
        elif select == 2:
            with open(default, "a") as f:
                content = input("Write a new note: ")
                f.write(content+":::"+time.strftime("%X %x")+"\n")
        elif select == 3:
            with open(default, "w") as f:
                content = ""
                f.write(content)
                print("Notes deleted.")
        elif select == 4:
            name = input("Give the name of the new file: ")
            getfile(name)
            default = name
            
            
        continue 
    print("Notebook shutting down, thank you.")
    
if __name__ == "__main__":
    main()
```
